```python
import pandas as pd
import matplotlib.pyplot as plt

# Load
sheet_key = "1ocBuWqHA4e9m2-qyEMvfWqed6SUOEHlj_c2GenObysc"
url = f"https://docs.google.com/spreadsheets/d/{sheet_key}/export?format=csv"
df = pd.read_csv(url)

# Ensure year column is numeric
df["year"] = pd.to_numeric(df["year"], errors="coerce")

# ------------------- 1. Top Individuals -------------------
top_individuals = (
    df.groupby("Employee Name")["New_Award_title"]
      .count()
      .reset_index(name="Total Awards")
      .sort_values("Total Awards", ascending=False)
)

# ------------------- 2. Top Teams -------------------
top_teams = (
    df.groupby("Team name")["New_Award_title"]
      .count()
      .reset_index(name="Total Awards")
      .sort_values("Total Awards", ascending=False)
)

# ------------------- 3. Award Title Popularity -------------------
award_popularity = (
    df.groupby("New_Award_title")["Employee Name"]
      .count()
      .reset_index(name="Award Count")
      .sort_values("Award Count", ascending=False)
)

# ------------------- 4. Year-wise Award Count -------------------
year_wise = (
    df.groupby("year")["New_Award_title"]
      .count()
      .reset_index(name="Total Awards")
      .sort_values("year")
)

print("\nTop Individuals:")
print(top_individuals.head(10))

print("\nTop Teams:")
print(top_teams.head(10))

print("\nMost Frequent Award Titles:")
print(award_popularity.head(10))

print("\nYear-wise Award Distribution:")
print(year_wise)

# ------------------- 5. Visualization -------------------

# Top Individuals
plt.figure(figsize=(9,5))
plt.barh(top_individuals.head(10)["Employee Name"], top_individuals.head(10)["Total Awards"])
plt.xlabel("Number of Awards")
plt.ylabel("Employee Name")
plt.title("Top 10 Employees by Awards")
plt.show()

# Top Teams
plt.figure(figsize=(9,5))
plt.barh(top_teams.head(10)["Team name"], top_teams.head(10)["Total Awards"])
plt.xlabel("Number of Awards")
plt.ylabel("Team Name")
plt.title("Top Teams by Awards")
plt.show()

# Award Title Popularity
plt.figure(figsize=(9,5))
plt.barh(award_popularity.head(10)["New_Award_title"], award_popularity.head(10)["Award Count"])
plt.xlabel("Count")
plt.ylabel("Award Title")
plt.title("Most Popular Award Titles")
plt.show()

# Year-wise Trend
plt.figure(figsize=(9,5))
plt.plot(year_wise["year"], year_wise["Total Awards"])
plt.xlabel("Year")
plt.ylabel("Number of Awards")
plt.title("Awards Trend Over Years")
plt.show()
```

```python
manager_influence = (
    df.groupby("Nominated By")["New_Award_title"]
      .count()
      .reset_index(name="Total Nominations")
      .sort_values("Total Nominations", ascending=False)
)

print(manager_influence.head(10))

plt.figure(figsize=(9,5))
plt.barh(manager_influence.head(10)["Nominated By"], manager_influence.head(10)["Total Nominations"])
plt.xlabel("Total Nominations")
plt.ylabel("Manager")
plt.title("Top Managers Influencing Recognition")
plt.show()
```

```python
award_types = (
    df.groupby("New_Award_title")["Employee Name"]
      .count()
      .reset_index(name="Award Count")
      .sort_values("Award Count", ascending=False)
)

print(award_types)

plt.figure(figsize=(9,5))
plt.barh(award_types["New_Award_title"], award_types["Award Count"])
plt.xlabel("Count")
plt.ylabel("Award Title")
plt.title("Which Award Types are Given Most?")
plt.show()
```

```python
month_trend = (
    df.groupby(["year", "Month"])["New_Award_title"]
      .count()
      .reset_index(name="Total Awards")
      .sort_values(["year", "Month"])
)

print(month_trend)

plt.figure(figsize=(10,5))
plt.plot(month_trend["Month"], month_trend["Total Awards"])
plt.xlabel("Month")
plt.ylabel("Number of Awards")
plt.title("Seasonal Award Trend Across Months")
plt.show()
```

```python
team_awards = (
    df.groupby("Team name")["New_Award_title"]
      .count()
      .reset_index(name="Total Awards")
      .sort_values("Total Awards", ascending=False)
)

print(team_awards)

plt.figure(figsize=(9,5))
plt.barh(team_awards["Team name"], team_awards["Total Awards"])
plt.xlabel("Total Awards")
plt.ylabel("Team Name")
plt.title("Recognition Distribution Across Teams")
plt.show()
```

```python
individual_award_counts = (
    df.groupby("Employee Name")["New_Award_title"]
      .count()
      .reset_index(name="Total Awards")
)

# Classify employees
individual_award_counts["Category"] = pd.cut(
    individual_award_counts["Total Awards"],
    bins=[0,1,3,10,100],
    labels=["One-Time Winner", "Occasional Performer", "Frequent Winner", "Star Performer"]
)

print(individual_award_counts.sort_values("Total Awards", ascending=False).head(15))

category_summary = individual_award_counts["Category"].value_counts()
print("\nCategory Summary:\n", category_summary)

plt.figure(figsize=(6,6))
plt.pie(category_summary, labels=category_summary.index, autopct='%1.1f%%')
plt.title("Employee Recognition Distribution by Consistency")
plt.show()
```
